## **Data Cleaning**

### 1. Đổi tên và Loại bỏ cột

Đổi tên cột bằng cách thay thế dấu "_" bằng khoảng trắng và viết hoa các từ.

```python
# Rename columns
data.columns = data.columns.str.replace('_',' ')
data.columns = data.columns.str.title()
data.columns
```

Loại bỏ các cột không cần thiết.

```python
# Drop columns
data = data.drop(columns=["Title", "Video Views For The Last 30 Days",
                          "Subscribers For Last 30 Days", "Latitude", "Longitude", "Abbreviation"])
```

### 2. Loại bỏ ký tự lạ và Trùng lặp của cột Youtuber

Loại bỏ các ký tự lạ trong danh sách.

```python
pattern = r'[^a-zA-Z0-9\s.,!?&\'-]'
```

```python
# Remove special characters
data['Youtuber'] = data['Youtuber'].apply(lambda x: re.sub(pattern, '', x))
data['Youtuber'] = data['Youtuber'].str.strip()
data['Youtuber'] = data['Youtuber'].str.lstrip('- ')
```

```python
data.tail(10)
```

Loại bỏ các dòng dữ liệu chứa giá trị rỗng trong cột Youtuber.

```python
# Check Youtuber column after remomving characters
blanks_data_youtuber = data[data['Youtuber'] == '']
blanks_data_youtuber
```

```python
# Remove those blank rows
data = data.drop(blanks_data_youtuber.index)
data.reset_index(drop= True, inplace= True)
```

Xóa bỏ trùng lặp.

```python
# Print rows that youtubers having duplicates
youtubers = data['Youtuber'].value_counts()
duplicate_youtubers = youtubers[youtubers > 1].index.tolist()

for youtuber in duplicate_youtubers:
    print(data[data['Youtuber'] == youtuber])
```

**Ghi chú:** Có sự trùng lặp, giữ lại Youtuber có ngày tạo kênh gần nhất

```python
# Remove rank values = 220 or 915
data = data[~data['Rank'].isin([220, 915])]
```

```python
# Check for duplicate rows
duplicate_rows = data[data.duplicated()]

# Print the number of duplicate rows
print(f"Number of duplicates: {len(duplicate_rows)}")

# Print the duplicate rows
print(duplicate_rows)
```

#### Biến số

##### Video Views Rank, Country Rank, Channel Type Rank

```python
# Replace null of Video Views Rank, Country Rank, Channel Type Rank with 0
# and replace type data to int
data['Video Views Rank'] = data['Video Views Rank'].fillna(0).astype(int)
data['Country Rank'] = data['Country Rank'].fillna(0).astype(int)
data['Channel Type Rank'] = data['Channel Type Rank'].fillna(0).astype(int)
```

##### Created Year, Created Month, Created Date

```python
# Print rows that Created Year is null
null_created_year = data[data['Created Year'].isnull()]
null_created_year
```

**Ghi chú:** Created Year, Created Month, Created Date đều có 5 giá trị thiếu và trên cùng 5 dòng => chỉ cần xóa bỏ giá trị thiếu ở Created Year.

```python
# Remove null of Created Year column
data = data.dropna(subset=['Created Year'])

# Change type float64 to int64
data['Created Year'] = data['Created Year'].astype('int64')
```

##### Gross Tertiary Education Enrollment, Unemployment Rate, Population, Urban Population

```python
# Change data type of Unemployment Rate to float64
data["Unemployment Rate"] = data["Unemployment Rate"].astype("float64")
```

Tính toán các giá trị trung bình.

```python
# Data without null
data_not_null = data.dropna(subset= ['Gross Tertiary Education Enrollment (%)',
                                     'Unemployment Rate', 'Population', 'Urban Population'])

# Mean Education Enrollment
mean_edu = round(data_not_null['Gross Tertiary Education Enrollment (%)'].mean(),1)
print('Mean Gross Tertiary Education Enrollment:', mean_edu)

# Mean Unemployment Rate
mean_unemp = round(data_not_null['Unemployment Rate'].mean(),1)
print('Mean Unemployment Rate:', mean_unemp)

# Mean Population
mean_popu = round(data_not_null['Population'].mean(),0)
print('Mean Population:', mean_popu)

# Mean Urban Population
mean_urban = round(data_not_null['Urban Population'].mean(),0)
print('Mean Urban Population:', mean_urban)
```

Thay thế giá trị thiếu bằng giá trị trung bình.

```python
# Replace null with mean
data['Gross Tertiary Education Enrollment (%)'] = data['Gross Tertiary Education Enrollment (%)'].fillna(mean_edu)
data['Unemployment Rate'] = data['Unemployment Rate'].fillna(mean_unemp)
data['Population'] = data['Population'].fillna(mean_popu)
data['Urban Population'] = data['Urban Population'].fillna(mean_urban)
```

```python
# Fill same values on the same rows of Channel Type to Category if Category is null
data['Category'] = data.apply(lambda row: row['Channel Type'] if pd.isna(row['Category'])
                              else row['Category'], axis=1)

# Fill same values on the same rows of Category to Channel Type if Channel is null
data['Channel Type'] = data.apply(lambda row: row['Category'] if pd.isna(row['Channel Type'])
                                  else row['Channel Type'], axis=1)
```

```python
## Howto & Style => Howto, Gaming => Games, Film & Animation => Film, People & Blogs => People
# Shows => Entertainment, Pet & Animals => Animals, Science & Technology => Tech
# Replace values in 'Channel Type' column
data['Channel Type'] = data['Channel Type'].replace({'Howto & Style': 'Howto', 'Gaming': 'Games',
                                                     'Film & Animation': 'Film', 'People & Blogs': 'People','Shows': 'Entertainment',
                                                     'Pets & Animals': 'Animals', 'Science & Technology': 'Tech'})
```

```python
# Create a new column named Region
data['Region'] = 'Unknown'

# Set Region to 'Asia' for specific countries
asia_countries = ['India', 'Japan', 'Pakistan', 'Philippines', 'Thailand', 'Afghanistan', 'Bangladesh',
                  'Indonesia', 'Turkey', 'China', 'Iraq', 'Egypt', 'Vietnam', 'Malaysia', 'Singapore', 'South Korea']
data.loc[data['Country'].isin(asia_countries), 'Region'] = 'Asia'

# Set Region to 'Europe' for specific countries
europe_countries = ['Russia', 'United Kingdom', 'Netherlands', 'Sweden', 'Spain', 'Andorra',
                    'Latvia', 'Switzerland', 'Germany', 'France', 'Italy', 'Finland', 'Ukraine']
data.loc[data['Country'].isin(europe_countries), 'Region'] = 'Europe'

# Set Region to 'America' for specific countries
america_countries = ['United States', 'Canada', 'Mexico', 'Barbados',
                     'Brazil', 'Argentina', 'Chile', 'Cuba', 'El Salvador',
                     'Colombia', 'Venezuela', 'Ecuador', 'Peru', 'Bolivia',
                     ]
data.loc[data['Country'].isin(america_countries), 'Region'] = 'America'

# Set Region to 'Africa' for specific countries
africa_countries = ['Morocco', 'United Arab Emirates', 'Kuwait', 'Jordan', 'Saudi Arabia']
data.loc[data['Country'].isin(africa_countries), 'Region'] = 'Africa'

# Set Region to 'Oceania' for specific countries
oceania_countries = ['Australia', 'Samoa']
data.loc[data['Country'].isin(oceania_countries), 'Region'] = 'Oceania'

# Print the updated DataFrame
data[['Rank', 'Country', 'Region']].head(10)
```

### Youtube Creator Awards

```python
data['Youtube Creator Awards'] = ''
for index, row in data.iterrows():
    if row['Subscribers'] >= 100000000:
        data.at[index, 'Youtube Creator Awards'] = 'Red Diamond'
    elif 10000000 <= row['Subscribers'] < 100000000:
        data.at[index, 'Youtube Creator Awards'] = 'Diamond'
    elif 1000000 <= row['Subscribers'] < 10000000:
        data.at[index, 'Youtube Creator Awards'] = 'Gold'
    elif 100000 <= row['Subscribers'] < 1000000:
        data.at[index, 'Youtube Creator Awards'] = 'Silver'
    else:
        data.at[index, 'Youtube Creator Awards'] = 'Other'
```
